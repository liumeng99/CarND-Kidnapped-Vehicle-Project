```python
import numpy as np
from numpy import genfromtxt
gt_data = genfromtxt(
    'data/gt_data.txt',
    delimiter=" ",
    skip_header=0,
    dtype=[
            ('x', 'f8'),
            ('y', 'f8'),
            ('a', 'f8')
    ])

filter_data = genfromtxt(
    'out.txt',
    delimiter=" ",
    skip_header=0,
    dtype=[
            ('x', 'f8'),
            ('y', 'f8'),
            ('a', 'f8')
    ])

# print (data['px_state'])

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16, 8))
plt.plot(gt_data ['x'], gt_data ['y'], filter_data ['x'], filter_data ['y'])
```
